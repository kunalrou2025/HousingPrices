```python
def factorial(value):
    fact = 1
    for i in range(1, value + 1):
        fact *= i
    return fact

def combination(n, r):
    return factorial(n) // (factorial(r) * factorial(n - r))

n = 10
r = 6
C = combination(n, r)
print("Combination: " + str(C))
```

```python
import pandas as pd
import numpy as np

# Read the CSV file
Raw_housing_data = pd.read_csv("1. Regression - Module - (Housing Prices).csv")

# Create the pivot table
zipcode_condition_sale3 = pd.pivot_table(
    Raw_housing_data,
    index=["Zipcode"],
    columns=['Condition of the House'],
    values=['Sale Price'],
    aggfunc=np.mean
)

# Display the result
print(zipcode_condition_sale3)
```

```python
import pandas as pd

# Assuming data is already loaded
# Define the function to classify luxury houses
def luxury_house(row):
    count = 0
    if row['WaterFront View'] == 'Yes':
        count += 1
    if row['Condition of the House'] in ["Good", "Excellent"]:
        count += 1
    if row['Overall grade'] >= 8:
        count += 2
    if count >= 2:
        return 'Yes'
    else:
        return 'No'

data['luxury_house'] = data[['WaterFront View', 'Condition of the House', 'Overall grade']].apply(luxury_house, axis=1)
print(data['luxury_house'].unique())

    
    
    
    
```
